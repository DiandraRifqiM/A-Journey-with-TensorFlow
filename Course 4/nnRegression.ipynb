##### Neural Network Regression w/ Tensor

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

Creating data to view it

```python
# Create feature
X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0, 17.0])

# Create label
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0, 27.0])

# Visualize
plt.scatter(X, y, edgecolors="blue", c="cyan")
```

```python
# Convert array to tensor
X_tensor = tf.constant(X)
y_tensor = tf.constant(y)

X_tensor, y_tensor
```

##### Steps in modelling w/ Tensorflow

- Create a model
- Complie the model
- Fit the model

```python
X_tensor = np.expand_dims(X, axis=-1)
y_tensor = np.expand_dims(y, axis=-1)

# Set seed
tf.random.set_seed(42)

# Create model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

model.fit(X_tensor, y_tensor, epochs=5)
```

```python
# Try to predict 
yPred = model.predict(tf.constant([[17.0]]))
yPred
```

```python
yPred + 17.0
```

##### Evaluate model

Basic model workflow for NN

Create -> Compile -> Fit -> Predict -> Tweak -> Fit -> Predict -> Tweak -> Fit -> Evaluate

When comes to evaluate make sure to remember this important thing

VISUALIZE!!!

Visualize help us to see how close the predict data into the factual data, evaluatig model not just to print out the number of predicting accuracy but also to visualize how good our model to predict the data.

```python
# Make bigger dataset
X = tf.constant(tf.range(-100, 100, 4))
X = tf.expand_dims(X, axis=-1)
# X
```

```python
# Create labels
y = X+10
# y
```

```python
# Visualize
plt.scatter(X, y, color="cyan", edgecolors="blue")
```

The 3 Sets of Data

Training Set - Typically the model learn for about 70-80% of the total dataset.

Validation Set - The model get tuned on this set, usually they use 10-15% of the data.

Test Set - The model gets evaluated, usually they print out the predict accuracy or learned accuracy model same as validation they usually use 10-15% of the data left.

```python
len(X)
```

```python
# Create train data (80% of X and y)
X_train = X[:40]
y_train = y[:40]

len(X_train), len(y_train)
```

```python
# Create test data (20% of X and y)
X_test = X[40:]
y_test = y[40:]

len(X_test), len(y_test)
```

```python
# Visualized (Blue Train, Green Test)

plt.scatter(X_train, y_train, color="skyblue", edgecolors="blue", label="Test Data")
plt.scatter(X_test, y_test, color="cyan", edgecolors="green", label="Train Data")
plt.legend()
```

```python
# Create model

# Set Seed
tf.random.set_seed(42)

# Create Model
modelC = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation=None),
    tf.keras.layers.Dense(1)
])

# Compile Model
modelC.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
               metrics=["mae"])

# Fit Model
modelC.fit(X_train, y_train, epochs=100, verbose=2)
```

```python
modelC.summary(),
```

```python
# Create model

# Set Seed
tf.random.set_seed(42)

# Create Model
modelD = tf.keras.Sequential([
    tf.keras.layers.Dense(5, input_shape=[1], name="input_layer"),
    tf.keras.layers.Dense(1, name="output_layer")
])

# Compile Model
modelD.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
               metrics=["mae"])

# Fit Model
modelD.fit(X_train, y_train, epochs=100, verbose=2)
```

```python
modelD.summary()
```

What are does ?

- Total params: The total of every parameters in the entire model
- Trainable params: The updated parameters by the optimizer during training to minimize loss.
- Non-trainable params: Fixed parameters that don't change, often from frozen layers or batch normalization statistics.
- Extra parameters (like momentum or moving averages) that the optimizer tracks to manage the learning process.

##### Plot Model

```python
# Visualize model

from tensorflow.keras.utils import plot_model
# plot_model(model=modelD, show_shapes=True)
```

```python
# Visualize model prediction

yPredD = modelD.predict(X_test)

# Predict plot func
def predict_plot(train_data=X_train,
                 train_labels=y_train,
                 test_data=X_test,
                 test_labels=y_test,
                 predict=yPredD):

    plt.figure(figsize=(10, 7))

    plt.title("Model Prediction Plot")
    # Train plot 
    plt.scatter(train_data, train_labels, c="skyblue", edgecolors="blue", label="Train Data")
    # Test plot
    plt.scatter(test_data, test_labels, c="cyan", edgecolors="green", label="Test Data")
    # Predict plot
    plt.scatter(test_data, predict, c="salmon", edgecolors="red", label="Prediction Data")
    plt.legend()

```

```python
predict_plot(X_train, y_train, X_test, y_test, yPredD)
```

##### Model Evaluate

Depending on what kind of model you working on, there will be different metrics value to evaluate your model.

Since we're working on regression model, we'll be using MAE and MSE:

- MAE: Mean Absolute Error mean on average, how wrong is each of my model's predictions.
- MSE: Mean Square Error mean we square the average error we got 

```python
# Evaluate model

modelD.evaluate(X_test, y_test)
```

```python
# Calculate MAE

mae = tf.metrics.mae(y_true=y_test, y_pred=yPredD)
mae = np.mean(mae)
mae
```

```python
# Calculate MSE
mse = tf.metrics.mse(y_true=y_test, y_pred=yPredD)
mse = np.mean(mse)
mse
```

##### Running Experiments to Improve the Model

Build a model -> Fit -> Evaulate -> Tweak -> Fit -> Evaluate -> Tweak -> Fit -> Evaluate...

1. Get more data
2. Make your model larger
3. Train for longer

In this case we do 3 modelling experiments:
- model_1   : Same as the original model, 1 layer & 100 epochs
- model_2   : 2 layers & 100 epochs
- model_3   : 2 layers & 500 epochs

Model_1

```python
# Set random set
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

model_1.fit(X_train, y_train, epochs=100, verbose=2)
```

```python
# Model_1 predict & plot
yPred_1 = model_1.predict(X_test)
predict_plot(X_train, y_train, X_test, y_test, yPred_1)
```

```python
# Evaluate model_1
mae_1 = tf.keras.metrics.mae(yPred_1, y_test)
mae_1 = np.mean(mae_1)

mse_1 = tf.keras.metrics.mse(yPred_1, y_test)
mse_1 = np.mean(mse_1)

mae_1, mse_1
```

Model_2

```python
# Set seed
tf.random.set_seed(42)

# Create model
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

model_2.fit(X_train, y_train, epochs=100, verbose=2)
```

```python
# Predict and plot model_2
yPred_2 = model_2.predict(X_test)
predict_plot(X_train, y_train, X_test, y_test, yPred_2)
```

```python
# Evaluate mae & mse model_2

mae_2 = tf.metrics.mae(tf.squeeze(yPred_2), tf.squeeze(y_test))
mse_2 = tf.metrics.mse(tf.squeeze(yPred_2), tf.squeeze(y_test))

mae_2, mse_2
```

Model_3

```python
# Set seed
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])


model_3.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

model_3.fit(X_train, y_train, epochs=500, verbose=2)
```

```python
# Predict and plot model_3
yPred_3 = model_3.predict(X_test)
predict_plot(X_train, y_train, X_test, y_test, yPred_3)
```

```python
mae_3 = tf.metrics.mae(tf.squeeze(yPred_3), tf.squeeze(y_test))
mse_3 = tf.metrics.mse(tf.squeeze(yPred_3), tf.squeeze(y_test))

mae_3, mse_3
```

##### Comparing Each Model Performance

```python
import pandas as pd

model_results = [["Model_1", mae_1, mse_1],
                 ["Model_2", mae_2.numpy(), mse_2.numpy()],
                 ["Model_3", mae_3.numpy(), mse_3.numpy()]]

modelsCompare = pd.DataFrame(model_results, columns=["Model", "MAE", "MSE"])

modelsCompare
```

```python
# MAE Models Comparison

plt.bar(modelsCompare["Model"], modelsCompare["MAE"], color="cyan")
plt.title("Mean Absolute Error Models Comparison")
```

```python
# MSE Models Comparison

plt.bar(modelsCompare["Model"], modelsCompare["MSE"], color="salmon")
plt.title("Mean Absolute Error Models Comparison")
```

MAE & MSE Function

```python
def mae(y_pred, y_true):
    print(tf.metrics.mae(tf.squeeze(y_pred), tf.squeeze(y_true)))

def mse(y_pred, y_true):
    tf.metrics.mse(tf.squeeze(y_pred), tf.squeeze(y_true))
```
