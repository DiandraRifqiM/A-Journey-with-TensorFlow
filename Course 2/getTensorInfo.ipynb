##### Getting Information from Tensor

When dealing with tensors you probably want to be aware of the following attributes:
- Shape
- Rank
- Axis or Dimension
- Size

```python
import tensorflow as tf
```

```python
# Creating rank 4 dimension

rank4_tensor = tf.ones(shape=[2,2,4,5])
rank4_tensor
```

```python
rank4_tensor.ndim, rank4_tensor.shape, tf.size(rank4_tensor)
2*2*4*5
```

```python
print(f"Number of dimension previous tensor = {rank4_tensor.ndim}\nShape of tensor = {rank4_tensor.shape}\nSize of the tensor = {tf.size(rank4_tensor).numpy()}\
    \nElement along the 0 axis = {rank4_tensor.shape[0]}\nElement along the 5 axis = {rank4_tensor.shape[-1]}")
```

##### Slicing Tensor

```python
some_list = [1,2,3,4]
some_list[:2]
```

```python
# Get the first 2 element from tensor
rank4_tensor[:2, :2, :2, :2]
```

```python
# Get the first elemet from tensor except for the final 1
rank4_tensor[:1, :1, :1, :]
```

```python
rank2_tensor = tf.constant([[3,4],[2,5]])
rank2_tensor, rank2_tensor.shape, rank2_tensor.ndim
```

```python
# Get the last element from rank2 tensor
rank2_tensor[:, -1]
```

```python
# Add in extra dimension to tensor
rank3_tensor = rank2_tensor[..., tf.newaxis]
rank3_tensor
```

```python
# Another way to do it
tf.expand_dims(rank2_tensor, axis=-1)
```

```python
# Expand 0-axis
tf.expand_dims(rank2_tensor, axis=0)
```
