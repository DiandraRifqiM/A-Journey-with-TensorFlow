##### Tensor Manipulation (tensor operations)
Basic Operation

" + - * / "

```python
import tensorflow as tf
```

```python
# Addition
tensor = tf.constant([[12, 4],[8, 15]])
tensor + 10, tensor
```

```python
# Tensor saved addition
tensor1 = tensor + 10
tensor1
```

```python
# Subtraction
tensor2 = tensor - 2
tensor2
```

```python
# Multiplication
tensor * 2
```

```python
# Use build-in function
tensor3 = tf.multiply(tensor, 2)
tensor3
```

```python
# Division
tensor4 = tensor / 2
tensor4
```

##### Matrix Multiplication

2 Rules of Matrix Multiplication:
1. The inner dimension must match
2. The resulting matrix has the shape of the inner dimension

```python
# Matrix multiplication w/ build-in tf function
tensorA = tf.matmul(tensor1, tensor3)
tensorA
```

```python
# Matrix multiplication w/ python operator

tensorB = tensor2 @ tensor3
tensorB
```

```python
# Matrix multiplication problem
X = tf.constant([[4, 4],
                 [3, 5],
                 [1, 4]])
Y = tf.constant([[2, 4],
                 [1, 2],
                 [5, 5]])

X, Y
```

```python
tf.matmul(X, Y)
```

```python
# Fixed problem

X_reshape = tf.reshape(X, shape=(-1, 3))
tf.matmul(Y, X_reshape)
```

```python
# Transpose
tf.transpose(X_reshape)
```

```python
tf.matmul(tf.transpose(Y), X)
```

Note

You can multiply matrix by using:
1. tf.matmul()
2. tf.tensordot()

##### Changing Datatype of a Tensor

```python
B = tf.constant([5, 7])
B.dtype
```

```python
C = tf.constant([3.14, 1.61])
C.dtype
```

```python
# Change int32/float32 to int16/float16 (reduced precision)

D = tf.cast(B, dtype=tf.int16)
D, D.dtype
```

```python
E = tf.cast(C, dtype=tf.float16)
E
```

##### Tensor Aggregation

Aggregating Tensor = condensing tensor multiple values down to a smaller values

```python
# Get the absolute value
D = tf.constant([13, 12])
```

```python
# Get the absolute value
tf.abs(D)
```

Form of Aggregration

- Get the minimum
- Get the maximum
- Get the mean
- Get the sum

```python
# Tensor with a random value between 0 - 100 with the size of 50

import numpy as np
E = tf.constant(np.random.randint(0, 100, size=50))
E.shape, E.ndim, tf.size(E), E
```

```python
# Get the min val of tensor
tf.reduce_min(E)
```

```python
# Get the max val of tensor
tf.reduce_max(E)
```

```python
# Get the mean of tensor
tf.reduce_mean(E)
```

```python
# Get the sum of tensor
tf.reduce_sum(E)
```

🛠️ Exercise: find the variance and standard deviation of E

```python
# Get variance
tf.math.reduce_variance(tf.cast(E, dtype=tf.float16))
```

```python
# Get standard deviation
tf.math.reduce_std(tf.cast(E, dtype=tf.float16))
```

##### Find the Positional Max & Min Val

```python
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find positional max val
tf.argmax(F)
```

```python
# Get index positional max val
F[tf.argmax(F)]
```

```python
# Get the positional min val
tf.argmin(F)
```

```python
F[tf.argmin(F)]
```
