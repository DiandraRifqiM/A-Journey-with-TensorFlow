##### Tensor x NumPy


```python
import tensorflow as tf
import numpy as np
```

```python
# Create Numpy Through Tensor
A = tf.constant(np.array([3., 4., 5.]))
B = tf.constant([3., 4., 5.,])
A
```

```python
# Convert Tensor to Numpy
np.array(A), type(np.array(A))
```

```python
# Convert Tensor to Numpy
A.numpy(), type(A.numpy())
```

```python
A.dtype, B.dtype
```
