##### One Hot Encoding

```python
import tensorflow as tf
A = [1, 0, 0, 1]
A_encoded = tf.one_hot(A, depth=4)
A_encoded
```

```python
A_encoded2 = tf.one_hot(A, depth=4, on_value="Male", off_value="Female")
A_encoded2
```

Trying more Math Func in tensor (Sqaure, root, log)

```python
B = tf.range(1,15)
```

```python
# Square
tf.square(B)
```

```python
# Root
tf.sqrt(tf.cast(B, dtype=tf.float32))
```

```python
# Log
tf.math.log(tf.cast(B, dtype=tf.float32))
```

```python
# Cummulative Sum
tf.math.cumsum(B)
```

```python
# Gamma
tf.math.igamma(tf.cast(B, dtype=tf.float32), tf.cast(B, dtype=tf.float32))
```

```python
tf.math.tan(tf.cast(B, dtype=tf.float32))
```
