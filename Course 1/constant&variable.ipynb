##### TF Constant

```python
import tensorflow as tf
```

```python
# Craete Scalar

scalar = tf.constant(69, dtype=tf.int32)
scalar
```

```python
# Create Vector
vector = tf.constant([12, 2], dtype=tf.int32)
vector
```

```python
# Custome dtype

matrix = tf.constant([[2.3, 4.5],
                     [3.3, 4.2]], dtype=tf.float32)

matrix.ndim
matrix.dtype
```

```python
# Create a tensor

tensor = tf.constant([[[4, 3, 5],
                       [2, 5, 7]],
                      [[3, 1, 2],
                       [6, 4, 3]],
                      [[4, 4, 3],
                       [2, 5, 3]]]) 
tensor
tensor.ndim
```

##### What we've created

- Scalar = A single number 
- Vector = A nunber with direction (1 dimension)
- Matrix = A 2 dimensional nunbers
- Tensor = an n-dimensional array of numbers (When n can be any number, a n=0 is scalar, a n=1 is vector)

##### TF Variable

```python
# Create Variable
changeable_tensor = tf.Variable([10, 15])
unchangeable_tensor = tf.constant([12, 17])

changeable_tensor, unchangeable_tensor
```

```python
# Assign variable
changeable_tensor[1].assign(23)
changeable_tensor
```
