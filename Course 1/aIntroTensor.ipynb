##### Rank on Tensor
Rank: Dimensional of number in tensor

```python
rank1_tensor = tf.Variable(["Hello", "Tensorflow"], tf.string)
rank2_tensor = tf.Variable([[3, 4, 5], [6, 7, 8]], tf.int64)
```

```python
tf.rank(rank2_tensor)
```

##### Modify Tensor Shape

```python
tensor1 = tf.ones([1,2,3])
tensor2 = tf.reshape(tensor1, [2, 3, 1])
tensor3 = tf.reshape(tensor2, [3, -1])
print(tensor1)
print(tensor2)
print(tensor3)
tensor2.ndim
```

##### Types of Dimension

- Scalar = A single number 
- Vector = A nunber with direction (1 dimension)
- Matrix = A 2 dimensional nunbers
- Tensor = an n-dimensional array of numbers (When n can be any number, a n=0 is scalar, a n=1 is vector)

```python
# Craete Scalar

scalar = tf.constant(69, dtype=tf.int32)
scalar
```

```python
# Create Vector
vector = tf.constant([12, 2], dtype=tf.int32)
vector
```

```python
# Custome dtype

matrix = tf.constant([[2.3, 4.5],
                     [3.3, 4.2]], dtype=tf.float32)

matrix.ndim
matrix.dtype
```

```python
# Create a tensor

tensor = tf.constant([[[4, 3, 5],
                       [2, 5, 7]],
                      [[3, 1, 2],
                       [6, 4, 3]],
                      [[4, 4, 3],
                       [2, 5, 3]]]) 
tensor
tensor.ndim
```
