##### Generate Tensor Random Number

```python
import tensorflow as tf
```

```python
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=[3, 2])
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=[3, 2])
random_2, random_1, random_2 == random_1
```

##### Shuffle a tensor

```python
not_shuffled = tf.constant([[4, 3, 2],
                            [3, 5, 2],
                            [4, 2, 5]])
not_shuffled
```

```python
shuffled = tf.random.shuffle(not_shuffled)
shuffled
```

```python
# tf.random.set_seed(42)
shuffled_seed = tf.random.shuffle(not_shuffled, seed=42)
shuffled_seed
```
